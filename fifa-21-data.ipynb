```python
import numpy as np
import pandas as pd

# Load the dataset
df = pd.read_csv("/kaggle/input/fifa-21-messy-raw-dataset-for-cleaning-exploring/fifa21 raw data v2.csv", sep=',')

import warnings
warnings.filterwarnings("ignore")
```

```python
# Display general information about the dataset
print(df.info())
```

```python
# Drop unneeded columns
columns_to_drop = ['Name', 'photoUrl', 'playerUrl']
df.drop(columns=columns_to_drop, inplace=True)
```

```python
# Rename 'LongName' column
df.rename(columns={'LongName': 'Name'}, inplace=True)
display(df.head(5))
```

```python
# Check for and handle duplicates
duplicate_count = df.duplicated().sum()
if duplicate_count > 0:
    print(f"There are {duplicate_count} duplicate rows in the DataFrame.")
    df.drop_duplicates(inplace=True)
    print("Duplicates removed.")
else:
    print("No duplicate rows found in the DataFrame.")
```

```python
# Display columns with missing values and their counts
missing_values = df.isnull().sum()
missing_columns = missing_values[missing_values > 0]
print("Columns with missing values and their counts:")
print(missing_columns)
```

```python
# Set 'ID' as the index
df.set_index('ID', inplace=True)
df.head()
```

```python
# Strip leading and trailing characters from object columns
object_columns = df.select_dtypes(include='object').columns
df[object_columns] = df[object_columns].apply(lambda x: x.str.strip() if x.dtype == 'object' else x)
```

```python
columns = list(df.columns)
pd.set_option('display.max_columns', 74)
display(df.head(5))
```

```python
# Changing Weight to numeric
mask = df['Weight'].astype(str).str.endswith('lbs')
df['Weight'] = df['Weight'].astype(str).str.strip('kglbs')
df['Weight'] = pd.to_numeric(df['Weight'], errors='coerce')  # Use 'coerce' to handle non-numeric values
df['Weight'] = np.where(mask, df['Weight'] * 0.45359237, df['Weight'])
```

```python
df['Weight'] = df['Weight'].astype(int)
df['Weight'].unique()
```

```python
import numpy as np

# Changing Height to numeric
mask_cm = df['Height'].astype(str).str.endswith('cm')
mask_inch = df['Height'].astype(str).str.endswith('"')
```

```python
# Remove units and convert to numeric
df['Height'] = df['Height'].astype(str).str.strip('cm"')
df['Height'] = pd.to_numeric(df['Height'], errors='coerce')
```

```python
# Use a constant for conversion factor
CM_TO_METER = 0.01
INCH_TO_METER = 0.0254

# Convert heights to meters, handling mixed units
df['Height'] = np.where(mask_cm, df['Height'] * CM_TO_METER, df['Height'])
df['Height'] = np.where(mask_inch, df['Height'] * INCH_TO_METER, df['Height'])
```

```python
# Convert 'Hits' to int
df['Hits'] = pd.to_numeric(df['Hits'].str.rstrip('K').fillna(0)) * 1000

# Convert 'Value' to int
df['Value'] = pd.to_numeric(df['Value'].str.replace('€', '').replace({'M': '*1000000', 'K': '*1000'}, regex=True).map(pd.eval))

# Convert 'Wage' to int
df['Wage'] = pd.to_numeric(df['Wage'].str.replace('€', '').replace({'M': '*1000000', 'K': '*1000'}, regex=True).map(pd.eval))

# Convert 'Release Clause' to int
df['Release Clause'] = pd.to_numeric(df['Release Clause'].str.replace('€', '').replace({'M': '*1000000', 'K': '*1000'}, regex=True).map(pd.eval))

# Convert 'W/F', 'SM', 'IR' to int
df[['W/F', 'SM', 'IR']] = df[['W/F', 'SM', 'IR']].astype(str).apply(lambda x: x.str.strip("★").astype(int))
```

```python
# Expand 'Contract' column into 'Contract Started' and 'Contract Ended'
contract_split = df[df['Contract'].str.contains('~')]['Contract'].str.split('~', expand=True).astype('Int64')
df[['Contract Started', 'Contract Ended']] = contract_split
```

```python
# Convert 'Joined' to datetime
df['Joined'] = pd.to_datetime(df['Joined'], format='%b %d, %Y')
```
